# Baseline model for RNN

## 1. Classifying bank complaints using

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras.models import Sequential 
from keras.layers import Flatten, Dense
from keras.preprocessing.text import Tokenizer
from keras.layers import Embedding, Dense, Dropout
from keras import preprocessing

df = pd.read_csv('Bank_complaints.csv')
df.head()
```

```python
import random
random.seed(123)
df = df.sample(2000)
```

```python
df.index = range(2000)
```

```python
product = df["Product"]
```

There are 7 types of complaints

```python
product.value_counts()
```

```python
df.info()
```

```python
complaints = df["Consumer complaint narrative"]
```

Only keep 1,000 most common words and use one-hot encoding

```python
n_words=1000
tokenizer = Tokenizer(num_words = n_words)

tokenizer.fit_on_texts(complaints)
complaints_encoded = tokenizer.texts_to_sequences(complaints)

one_hot_results= tokenizer.texts_to_matrix(complaints, mode='binary')
word_index = tokenizer.word_index
np.shape(one_hot_results)
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(product)
list(le.classes_)
product_cat = le.transform(product)
product_onehot = to_categorical(product_cat)
```

```python
np.shape(product_onehot)
```

```python
np.shape(one_hot_results)
```

```python
x_train, x_test, y_train, y_test = train_test_split(one_hot_results,  
            product_onehot, test_size=0.10, random_state=1234)
```

```python
random.seed(123)
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(50, activation='relu', input_shape=(1000,))) #2 hidden layers
model.add(layers.Dense(25, activation='relu'))
model.add(layers.Dense(7, activation='softmax'))
```

```python
model.compile(optimizer='SGD',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=300, batch_size=256, validation_split=0.2)
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
```
